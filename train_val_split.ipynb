```python
from sklearn.model_selection import train_test_split
from glob import glob
import os
import shutil
```

```python
# the dir of labels
labelme_path = "dataset\\annotation_yolo\\"
raw_img_path = 'dataset\\raw imgs\\'

isUseTest = False  # 是否创建test集

# get all json label files
files = glob(labelme_path + "*.txt")
files = [i.replace("\\", "/").split("/")[-1].split(".txt")[0] for i in files]
len(files)
files
```

```python
trainval_files = files # I don't use test set
train_files, val_files = train_test_split(trainval_files, test_size=0.02, random_state=55)
```

```python
print(len(train_files))
len(val_files)
```

## make dataset

```python
if not os.path.exists('tmp/'):
        os.makedirs('tmp/')

wd = os.getcwd()
wd
```

```python
'''
create a new dir called temp
store paths of train and val images respectively
'''
def make_temp(files, txt_Name):
    
    list_file = open('tmp/%s.txt' % (txt_Name), 'w') # txt_name: train / val / test
    
    for txt_file_ in files:
        imagePath = raw_img_path + txt_file_ + ".jpg" # raw image file path
        list_file.write('%s\\%s\n' % (wd, imagePath))
```

```python
make_temp(train_files, 'train')
make_temp(val_files, 'val')
```

```python
# make dataset
file_list = ['train', 'val']
for file in file_list:
    # make dirs if not exist
    if not os.path.exists('VOC/images/%s' % file):
        os.makedirs('VOC/images/%s' % file)
    if not os.path.exists('VOC/labels/%s' % file):
        os.makedirs('VOC/labels/%s' % file)

    # 保证tmp里面是否存在train, val文件夹
    assert(os.path.exists('tmp/%s.txt' % file))

    # 打开tmp中train, val
    f = open('tmp/%s.txt' % file, 'r')
    # 读取每个文件夹里面存放的图片地址
    lines = f.readlines()
    for line in lines:
        # Python join() 方法用于将序列中的元素以指定的字符连接生成一个新的字符串
        line = "/".join(line.split('\\')[-3:]).strip()
        print(line)
        shutil.copy(line, "VOC/images/%s" % file)
        label_path = 'dataset\\annotation_yolo'.replace('\\', '/')
        label_fname = line.split('/')[-1].replace('jpg', 'txt').strip()
        line = label_path + '/' + label_fname
        shutil.copy(line, "VOC/labels/%s/" % file)
```
